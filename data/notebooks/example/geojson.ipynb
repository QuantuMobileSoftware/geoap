```python
def _str_time_prop(start, end, format, prop):
    """Get a time at a proportion of a range of two formatted times.

    start and end should be strings specifying times formated in the
    given format (strftime-style), giving an interval [start, end].
    prop specifies how a proportion of the interval to be taken after
    start.  The returned time will be in the specified format.
    """

    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(format, time.localtime(ptime))


def random_date(start="2020-01-01T00:00:00", end=datetime.today().strftime('%Y-%m-%dT%H:%M:%S')):
    return _str_time_prop(start, end, '%Y-%m-%dT%H:%M:%S', random.random())
```

```python
def generate_features(): 
    polygons = [kharkiv_zoo, kharkiv_park, homilsha, hytor, osnova_lake]
    points = [point1, point2, point3, point4]
    lines = [line1, line2, line3, line4]

    geometries = [*polygons, *points, *lines]
    

    max_features = 10
    features_number = random.randint(1, max_features)
    print(f"{features_number} features will be in current file")
    
    data_template = [
        {
            "x": ["2019-04-02", "2019-04-07"],
            "y": [950, 3100],
            "type": "bar"
        }
    ];
    layout = {
        "title": "Cutting area changes",
        "yaxis": {
            "title": "Area, m²"
        },
        "xaxis": {
            "type": "category",
            "tickangle": -45
        }
    };

    features = list()
    for _ in range(0, features_number):
        
        # id_ = str(uuid.uuid4())        
        geometry = random.choice(geometries)
        
        data = copy.deepcopy(data_template)
        data[0]["y"] = [random.randint(1, 5000), random.randint(1, 10000)]
        style = dict(color=random.choice(["red", "green", "blue"]))
        label = random.choice(["test_label_1", "test_label_2", "test_label_3"])
        feature = Feature(geometry=geometry, properties=dict(style=style, data=data, layout=layout, label=label))
        features.append(feature)
    return features
```

```python
def generate_item():
    file_date = random_date() 
    id_ = str(uuid.uuid4())
    name = random.choice(["Red", "Green", "Blue", "NDVI", "NDWI", "PSRI", ])
    
    features = generate_features()
    print(f"Generated {len(features)} features")
    
    if len(features) == 1:
        item = features[0]
    else:
        item = FeatureCollection(features=features)
    
    item['id'] = id_
    if random.choice([True, False]):
        item['name'] = name
        print(f"Randomly selected to add name {name}")
    
    # if random.choice([True, False]):
    #    item['request_id'] = 1
    #    print(f"Randomly selected to add request_id {item['request_id']}")
    
    if random.choice([True, False]):             
        start_date = random_date()
        end_date = random_date(start_date)
        item['start_date'] = start_date
        item['end_date'] = end_date
        print(f"Randomly selected to add start_date: {start_date} and end_date: {end_date}")   
    return item, f"/home/{NB_USER}/work/results/example/{id_}_{name}_{file_date.split('T')[0]}.geojson.temp"
```

### main execution cell

```python
print("Generating .geojson file...")

item, temp_path = generate_item()
os.makedirs(os.path.dirname(temp_path), exist_ok=True)

with open(temp_path, 'w') as file:
    geojson.dump(item, file)
    
print(f"Generated temp file: {temp_path}")
output_path = temp_path[:-5]
print(f"Rename temp file: {temp_path} to {output_path}")
os.rename(temp_path, output_path)

print("Execution finished")
```
